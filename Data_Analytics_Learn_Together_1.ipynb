### Now I set up a sample matrix made of **Lists**
* This is like the multiple options my code can use to generate data 

```python
first_names = ['John', 'Jane', 'Mary', 'Michael', 'Sarah', 'David', 'Laura', 'Robert', 'Emily', 'James']
last_names = ['Smith', 'Johnson', 'Williams', 'Jones', 'Brown', 'Davis', 'Miller', 'Wilson', 'Moore', 'Taylor']
genders = ['Male', 'Female']
degrees = ['Computer Science', 'Engineering', 'Mathematics', 'Physics', 'Chemistry', 'Biology', 'Economics', 'Psychology']
ins_users = ['user1', 'user2', 'user3', 'user4', 'user5']
domains = ['hotmail.com', 'yahoo.com', 'gmail.com']
```

### It's time to generate the random data
* **Important** thing to notice, I wanted to practice how to generate emails by mixing different Lists

```python
data = {
    'first_name': [random.choice(first_names) for _ in range(100)],
    'last_name': [random.choice(last_names) for _ in range(100)],
    'gender': [random.choice(genders) for _ in range(100)],
    'age': np.random.randint(18, 30, size=100),
    'email': [],
    'mean_grades': np.round(np.random.uniform(6.0, 10.0, size=100), 2),
    'degree': [random.choice(degrees) for _ in range(100)],
    'ins_user': [random.choice(ins_users) for _ in range(100)],
}
```

#### I'd like to explain this because it was complicated for me to understand
* This line of code generates a list of email addresses and assigns it to the email column in the data dictionary. The email addresses are created by concatenating the lowercase versions of each student's first and last names, followed by a randomly chosen email domain.

* **data['email'] = [...]**
This assigns the list generated by the list comprehension to the email key in the data dictionary. This list will contain 100 email addresses.

##### **String Formatting**
* **data['first_name'][i].lower()**
  
data['first_name'][i] accesses the first name of the student at index i in the first_name list within the data dictionary.
.lower() converts the first name to all lowercase letters.

* **data['last_name'][i].lower()**

  
Similarly, data['last_name'][i] accesses the last name of the student at index i in the last_name list within the data dictionary.
.lower() converts the last name to all lowercase letters.

* **random.choice(domains)**
  
random.choice(domains) selects a random domain from the domains list, which contains "hotmail.com", "yahoo.com", and "gmail.com".

```python
data['email'] = [
    f"{data['first_name'][i].lower()}.{data['last_name'][i].lower()}@{random.choice(domains)}"
    for i in range(100)
]
```

### Finally it's time to create the data frame and Print it for visibility

```python
students_df = pd.DataFrame(data)

print(students_df)
```

#### What is the distribution of students by gender? Create a visualization showing the number of male and female students.

```python
# Let's find the number of male and female students and save it into a variable
male_students = len(students_df.loc[students_df['gender'] == 'Male'])
female_students =len(students_df.loc[students_df['gender'] == 'Female'])

# With my results from the count I save this into a Series
gender_counts = pd.Series({
    'Male': male_students,
    'Female': female_students
})

# I also want to show a written sentence about the info to give a punch to the bar chart
students_distro = 'The total count of students is {} Female and {} Male students, giving a total of {} students in the school'

# Using my series information I can now create a bar chart
gender_counts.plot(kind='bar', color=['blue', 'pink'])

# Let give format to my bar chart
plt.title('Distribution of students by gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')

# Now I show my plot and analysis 
plt.show()

# Printed Output
print(students_distro.format(gender_counts['Female'],
                             gender_counts['Male'],
                             gender_counts['Female'] + gender_counts['Male']))
                             
```

### **Grade Distribution:**
